# Задание 1

**Описание:** Создайте иерархию классов для разных типов сотрудников в компании. Реализуйте родительский класс Employee и дочерние классы Manager и Developer. Каждый класс должен иметь метод для расчета зарплаты на основе различных критериев класса.


Отрабатываемый принцип: Наследование



```python
class Employee:
  rank = 'Сотрудник'

  def __init__(self, name, hours, usd_per_hour):
    self.name = name
    self.hours = hours
    self.usd_per_hour = usd_per_hour

  def get_info(self):
    salary = self.hours * self.usd_per_hour
    return f'{self.rank} {self.name}, Зарплата: {salary}$ в день'

class Manager(Employee):
  rank = 'Менеджер'

class Developer(Employee):
  rank = 'Разработчик'

man = Manager('Петя', 8, 5)
print(man.get_info())

dev = Developer('Коля', 12, 1)
print(dev.get_info())
```

# Задание 2

**Описание:** Создайте иерархию классов для различных типов транспортных средств (Необходим один родительский класс и 3 дочерних). Реализуйте метод, который позволяет каждому транспортному средству возвращать собственное описание (Метод в каждом классе должен иметь одинаковое название). Продемонстрируйте вызов данного метода для каждого транспортного средства.


Отрабатываемый принцип: Полиморфизм

```python
class Vehicle:
  running_gear_type = 'ходовая'

  def __init__(self, name, max_speed, weight):
    self.name = name
    self.max_speed = max_speed
    self.weight = weight

  def __str__(self):
    return f'Транспортное средство: {self.name}. Максимальная скорость: {self.max_speed}км/ч.\
 Тип ходовой: {self.running_gear_type}. Масса: {self.weight}.'

class Tank(Vehicle):
  running_gear_type = 'гусеничная лента'

  def __init__(self, name, max_speed, weight, weapon_type):
    super().__init__(name, max_speed, weight)
    self.weapon_type = weapon_type

  def __str__(self):
    return f'Название танка: {self.name}. Максимальная скорость: {self.max_speed} км/ч.\
 Тип ходовой: {self.running_gear_type}. Масса: {self.weight}кг. Тип вооружения: {self.weapon_type}.'


class Bobrocar(Vehicle):
  def __init__(self, name, max_speed, weight, crew):
    super().__init__(name, max_speed, weight)
    self.crew = crew

  def __str__(self):
    return f'Название бобромобиля: {self.name}. Максимальная скорость: {self.max_speed} вёрст в час.\
 Масса: {self.weight}кг. Экипаж: {self.crew} отважных бобров.'

class Tractor(Vehicle):

  def __init__(self, name, max_speed, weight, thrust_force, running_gear_type):
    super().__init__(name, max_speed, weight)
    self.thrust_force = thrust_force
    self.running_gear_type = running_gear_type

  def __str__(self):
    return f'Название трактора: {self.name}. Максимальная скорость: {self.max_speed}км/ч.\
 Масса: {self.weight}. Сила тяги: {self.thrust_force}кН/кг. Тип ходовой: {self.running_gear_type}.'

bus = Vehicle('Автобус', 40, 3000)
tank = Tank('Т-2016', 60, 50000, 'Гаубица')
bober = Bobrocar('Бобродрайв - 3000', 240, 300, 10)
tractor = Tractor('Belarus', 34, 4000, 6, 'гусеничная лента')

print(bus)
print(tank)
print(bober)
print(tractor)
```

# Задание 3

Онлайн-магазин:
- Создайте модель для онлайн-магазина с классами Product, Order, Customer, и ShoppingCart.
- Product включает информацию о цене, наличии на складе и категории товара.
Order обрабатывает процесс покупки, включая расчет цены с учетом скидок и налогов.
- Customer управляет информацией о пользователе и его истории заказов.
- ShoppingCart позволяет добавлять, удалять и обновлять количество товаров перед оформлением заказа.


```python
class Product:
  def __init__(self, name, price, amount):
    self.name = name
    self.price = price
    self.amount = amount


class Customer:
  shopping_story = []
  def __init__(self, name):
    self.name = name

  def watch_story(self):
    print('Здравствуйте,', self.name)
    print('Ваша история покупок:')
    for purchare in self.shopping_story:
      print(purchare)


class Order:
  def __init__(self, name, discount, buy_amount): #discount в %, buy_amount в единицах товара
    self.name = name
    self.buy_amount = buy_amount
    self.discount = discount

  def buy_product(self, Product, Customer):
    if Product.amount >= self.buy_amount:
      Product.amount -= self.buy_amount
      total_cost = self.buy_amount * Product.price * (100 - self.discount) / 100
      check = f'Товар: {Product.name}. Количество: {self.buy_amount}.\
  Цена: {total_cost}. Скидка: {self.discount}% '
      Customer.shopping_story.append(check)
    else:
      print('Недостаточно товара')


client = Customer('Павел')
apple = Product('яблоко', 5, 10)
buy_apple = Order('яблоко', 5, 2)
bubble_gum = Product('жвачка', 10, 20)
buy_bubble_gum = Order('жвачка', 20, 15)

buy_apple.buy_product(apple, client)
buy_bubble_gum.buy_product(bubble_gum, client)
client.watch_story()
```

# Задание 4

Симулятор космического корабля:
- Создайте симулятор управления космическим кораблем с классами SpaceShip, CrewMember, и Mission.
- SpaceShip имеет атрибуты для управления топливом, состоянием корпуса, и текущей скоростью.
- CrewMember контролирует здоровье, навыки, и роли в команде (например, пилот, инженер).
- Mission определяет цели, ресурсы, и возможные события (например, аварии, встречи с астероидами).

```python
class SpaceShip:
  current_fuel = 0
  current_speed = 0
  ship_state = 100
  def __init__(self, spaceship_name, max_fuel, max_speed):
    self.spaceship_name = spaceship_name
    self.max_fuel = max_fuel
    self.max_speed = max_speed

  def __str__(self):
    return f'Название корабля: {self.spaceship_name}. Надёжность: {self.ship_state}.\
  Скорость: {self.current_speed} км/ч. Топливо в баке: {self.current_fuel} л'

  def add_fuel(self, fuel_amount):
    self.current_fuel += fuel_amount
    if self.current_fuel > self.max_fuel:
      self.current_fuel = self.max_fuel
      print('Топливо полностью заправлено')
    print(f'Заполненность: {self.current_fuel}/{self.max_fuel}л.')


class CrewMember:

  def __init__(self, role, SpaceShip):
    self.role = role
    self.ship = SpaceShip.spaceship_name

  def __str__(self):
    return f'Роль: {self.role}. Корабль: {self.ship}'

  def repair_ship(self, SpaceShip):
    if SpaceShip.ship_state <= 0:
      print('Нельзя починить сломанный корабль')
    elif self.role == 'Инженер':
      SpaceShip.ship_state += 5
      if SpaceShip.ship_state > 100:
        SpaceShip.ship_state = 100
      print('Надёжность корабля:', SpaceShip.ship_state)
    else:
      print('Чинить корабль может только инженер')

  def increase_speed(self, SpaceShip):
    if SpaceShip.ship_state <= 0:
      print('Корабль сломан, дальше не пойдёт. Конечная')
    elif SpaceShip.current_fuel < 20:
      print('Недостаточно топлива')
    elif self.role == 'Пилот':
      SpaceShip.current_speed += 5000
      SpaceShip.current_fuel -= 20
      if SpaceShip.current_speed > SpaceShip.max_speed:
        SpaceShip.current_speed = SpaceShip.max_speed
      print(f'Корабль движется со скоростью: {SpaceShip.current_speed}км/ч')
      print('Топливо в баке:', SpaceShip.current_fuel)
    else:
      print('Изменить скорость может только пилот')

  def decrease_speed(self, SpaceShip):
    if self.role == 'Пилот':
      SpaceShip.current_speed -= 5000
      if SpaceShip.current_speed < 0:
        SpaceShip.current_speed = 0
      print(f'Корабль движется со скоростью: {SpaceShip.current_speed}км/ч')
    else:
      print('Изменить скорость может только пилот')


class Missions:
  def __init__(self, incident, damage_to_ship, SpaceShip):
    self.incident = incident
    self.damage_to_ship = damage_to_ship

  def incident_ocure(self, SpaceShip):
    print(f'Произошло событие: {self.incident}. Урон кораблю: {self.damage_to_ship}')
    SpaceShip.ship_state -= self.damage_to_ship
    if SpaceShip.ship_state <= 0:
      print(f'"{SpaceShip.spaceship_name}" уничтожен. Члены экипажа в открытом космосе. ')
      SpaceShip.current_speed = 0
    else:
      print('Текущая надёжность корабля:', SpaceShip.ship_state)


ship = SpaceShip('Солярис', 100, 28000)
eng = CrewMember('Инженер', ship)
pilot = CrewMember('Пилот', ship)
ship.add_fuel(30)
pilot.increase_speed(ship)

print(ship)
pilot.increase_speed(ship)
pilot.increase_speed(ship)
print(eng)
print(pilot)
asteroid = Missions('Удар с астероидом', 120, ship)
asteroid.incident_ocure(ship)
eng.repair_ship(ship)
pilot.increase_speed(ship)






```

**Описание:** создайте консольную версию игры крестики-нолики, используя классы

```python
class Map:
  map = ['[ ]', '[ ]','[ ]' , '[ ]', '[ ]', '[ ]', '[ ]', '[ ]','[ ]' ]
  def show_map(self):
    for i in range(9):
      if (i+1) % 3 == 0:
        print(self.map[i])
      else:
        print(self.map[i], end = '')

  def check_for_win(self):
    if self.map[0] == self.map[1] == self.map[2] != '[ ]':
      return True
    elif self.map[3] == self.map[4] == self.map[5] != '[ ]':
      return True
    elif self.map[6] == self.map[7] == self.map[8] != '[ ]':
      return True
    elif self.map[0] == self.map[4] == self.map[8] != '[ ]':
      return True
    elif self.map[2] == self.map[4] == self.map[6] != '[ ]':
      return True
    else:
      return False

class Circle:
  visual = '[O]'
  turn = True

  def set_circle(self, Map, place):
    index = place - 1
    Map.map[index] = self.visual


class Cross:
  visual = '[X]'
  turn = False

  def set_cross(self, Map, place):
    index = place - 1
    Map.map[index] = self.visual


field = Map()
circ = Circle()
cross = Cross()

field.show_map()

turns = 0

while True:

  if circ.turn == True:
    square = int(input('Нолики, введите номер клетки: '))
    if square > 9 or square <= 0:
      print('Клеток всего 9')
      continue
    elif field.map[square - 1] != '[ ]':
      print('Эта клетка занята')
      continue
    else:
      circ.set_circle(field, square)
      circ.turn = False
      cross.turn = True
      if field.check_for_win() == True:
        field.show_map()
        print('Нолики победили!')
        break

    field.show_map()

    turns += 1
    if turns == 9:
      print('Ничья!')
      break

  else:
    square = int(input('Крестики, введите номер клетки: '))
    if field.map[square - 1] != '[ ]':
      print('Эта клетка занята')
      continue
    else:
      cross.set_cross(field, square)
      cross.turn = False
      circ.turn = True
      if field.check_for_win() == True:
        field.show_map()
        print('Крестики победили!')
        break

    field.show_map()

    turns += 1

```
