```python
numbersAmount = int(input("Введите количество чисел: "))
sum = 0
numb = 0
for i in range(numbersAmount):
  numb = float(input("Введите число: "))
  sum += numb
print("Среднее арифметическое: ", sum / numbersAmount)
```

Задание 7: Необходимо создать простой калькулятор, который позволяет пользователю выбрать одну из четырёх операций (+, -, *, /), ввести два числа и получить результат выполнения операции.

`Введите операцию:`

`Введите первое число:`

`Введите второе число:`

`Результат выполнения: `


```python
operation = input("Введите операцию: ")
num1 = int(input("Введите первое число: "))
num2 = int(input("Введите второе число: "))
if operation == '+':
  print("Результат выполнения: " , num1+num2)
elif operation == '-':
  print("Результат выполнения: " , num1-num2)
elif operation == '*':
  print("Результат выполнения: " , num1*num2)
elif operation == '/' or operation == ":":
  print("Результат выполнения: " , num1/num2)
```

# Практика 0.2.1

**Шахматы**

Даны стартовые и конечные координаты, а также фигура

Необходимо определить, может ли заданная фигура так ходить?

```python

vert1 = int(input("Введите начальную вертикаль фигуры: "))
hor1 = int(input("Введите начальную горизонталь фигуры: "))
vert2 = int(input("Введите конечную вертикаль фигуры: "))
hor2 = int(input("Введите конечную горизонталь фигуры: "))
pieceName = input("Введите название фигуры: ")

if pieceName == "Пешка":
  if vert1 == vert2 and abs(hor2 - hor1) == 1:
    print("Пешка может так ходить")
  elif vert1 == vert2 and abs(hor2 - hor1) == 2 and (hor1 == 2 or hor1 == 7):
    print("Пешка может так ходить")
  else:
    print("Пешка не может так ходить")
elif pieceName == "Король":
  if vert1 == vert2 and abs(hor1 - hor2) == 1:
    print("Король может так ходить")
  elif hor1 == hor2 and abs(vert1 - vert2) == 1:
    print("Король может так ходить")
  elif abs(vert1 - vert2) == 1 and abs(hor1-hor2) == 1:
    print("Король может так ходить")
  else:
    print("Король не может так ходить")
elif pieceName == "Слон":
  if abs(hor1 - hor2) == abs(vert1 - vert2):
    print("Слон может так ходить")
  else:
    print("Слон не может так ходить")
elif pieceName == "Ладья":
  if hor1 == hor2 or vert1 == vert2:
    print("Ладья может так ходить")
  else:
    print("Ладья не может так ходить")
elif pieceName == "Конь":
  if abs(vert1 - vert2) == 2 and abs(hor1 - hor2) == 1:
    print("Конь может так ходить")
  elif abs(vert1 - vert2) == 1 and abs(hor1 - hor2) == 2:
    print("Конь может так ходить")
  else:
    print("Конь не может так ходить")
elif pieceName == "Королева":
  if abs(hor1 - hor2) == abs(vert1 - vert2):
    print("Королева может так ходить")
  elif hor1 == hor2 or vert1 == vert2:
    print("Королева может так ходить")
  else:
    print("Королева не может так ходить")
```
