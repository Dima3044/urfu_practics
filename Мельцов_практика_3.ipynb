# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

***Нельзя использовать: len(), min(), max(), average(), sum()***

Ввод:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
nums_list = [int(x) for x in input().split()]
print(nums_list)
count = 0
sum = 0
max = nums_list[0]
for i in nums_list:
  sum += nums_list[count]
  if nums_list[count] > max:
    max = nums_list[count]
  count += 1
print("Сумма чисел: " , sum)
print("Среднее арифм.: " ,(sum/count))
print("Максимальное значение: " , max)
````

# **Задача 4.**


Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Ввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
import random

def game(botM,playerM):
  if playerM == "Камень":
    if botM == "Бумага":
      winner = "Компьютер"

    elif botM == "Ножницы":
      winner = "Игрок"

    elif botM == "Камень":
      winner = 'Дружба'



  elif playerM == "Бумага":
    if botM == "Бумага":
      winner = "Дружба"

    elif botM == "Ножницы":
      winner = "Компьютер"

    elif botM == "Камень":
      winner = 'Игрок'


  elif playerM == "Ножницы":
    if botM == "Бумага":
      winner = "Игрок"
    elif botM == "Ножницы":
      winner = "Дружба"
    elif botM == "Камень":
      winner = 'Компьютер'

  else:
    winner = "Ошибка"


  return winner




game_dict = {
    1: "Камень",
    2: "Ножницы",
    3: "Бумага"
}
botScore = 0
playerScore = 0

for i in range(3):
  print("Этап", i+1)
  playerMove = input("Игрок: ")
  botMove = game_dict[random.randint(1 , 3)]
  print("Компьютер:", botMove)
  game_result = game(botMove, playerMove)
  if game_result == "Игрок":
    playerScore += 1
    print("Победа: Игрок")
  elif game_result == "Компьютер":
    botScore += 1
    print("Победа: Компьютер")
  elif game_result == "Дружба":
    print("Ничья")
  else:
    print("Ошибка ввода")
    continue
  print()

if playerScore > botScore:
  print("Победил Игрок!")
else:
  print("Победил Компьютер!")
````

**Цифра на определенном месте:**

Последовательно записан натуральный ряд чисел.

Какая цифра стоит в N позиции


Вввод:

```
Введите номер позиции: 1234567890
```

Вывод:

```
8
```

````python
position = int(input())
count = position
tenDigit = 10
digitsPerNumber = 1

for i in range(1, count+1):
  if i == tenDigit:
    nums_str = ""
    position -= tenDigit * digitsPerNumber
    tenDigit *= 10
    digitsPerNumber += 1
  nums_str += str(i)
print(nums_str[position])
````
