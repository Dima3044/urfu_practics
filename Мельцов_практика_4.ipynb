**Задача 1:**



Интернет-магазин предлагает следующие условия скидок:

*   Для заказов больше 1000 единиц, клиент получает скидку 5%. Если клиент использует промокод SUPERDISCOUNT, он получает скидку 10% (вместо 5%).
*  Для заказов более 5000 единиц, клиент получает скидку 15%, а использование промокода SUPERDISCOUNT увеличивает скидку до 20% (вместо 15%).

Этап 1:
Ввод:
```
Введите стоимость единицы товара: 5
Введите количество товара: 1001
Введите промокод: GiVEMEDISCONT
```

Вывод:

```
Ваша скидка: 5%
Итоговая сумма: 4754.75
```
Этап 2:

Оформите ваш код в виде функции

````python
product_price = float(input("Введите стоимость единицы товара: "))
product_amount = int(input("Введите количество товара: "))
promo = input("Введите промокод: ")
disc = 0

if product_amount > 5000:
  if promo == "SUPERDISCOUNT":
    disc = 20
  else:
    disc = 15
elif product_amount > 1000:
  if promo == "SUPERDISCOUNT":
    disc = 10
  else:
    disc = 5

basicSum = product_price * product_amount
discountSum = basicSum * (disc / 100)
resultSum = basicSum - discountSum

print(f"Ваша скидка: {disc}%")
print(f"Итоговая сумма: {resultSum}")
````

> Алгоритм Евклида – это алгоритм нахождения наибольшего общего делителя (НОД) пары целых чисел.

Чуть подробнее, если вы забыли: [link](https://skysmart.ru/articles/mathematic/naibolshij-obshchij-delitel)


Ввод:
```
30 18
```

Вывод:

```

```

Этап 2:
Оформите ваш код в виде функции


````python
def EuclidAlg(num1, num2):
    if num1 > num2:
        biggerN = num1
        smallerN = num2
    else:
        biggerN = num1
        smallerN = num2

    remainder = biggerN % smallerN

    while remainder != 0:
        biggerN = smallerN
        smallerN = remainder
        remainder = biggerN % smallerN
    return smallerN
twoNums = input().split()
num1 = int(twoNums[0])
num2 = int(twoNums[1])
print(EuclidAlg(num1, num2))
````

Этап 2:

Ввод:
```
apple banana apple
```

Вывод:

```

```

````python
def RepeatsAmount(inputString):
    allWords = inputString.split()
    uniqueWords = set(allWords)

    for uniqueWord in uniqueWords:
        count = 0
        for word in allWords:
            if uniqueWord == word:
                count += 1
        print(f"{uniqueWord}: {count}")

inputStr = input()

RepeatsAmount(inputStr)
````

````python


def AnnoCheck(String):
    String = String.split(", ")

    firstS = String[0]
    secondS = String[1]

    lettersCount1 = 0
    lettersCount2 = 0

    for letter in firstS:
        if letter != ' ':
            lettersCount1 += 1
    for letter in secondS:
        if letter != ' ':
            lettersCount2 += 1
    #это чтобы учитывались только буквы, например при ape pl, el pap выдало true
    if lettersCount1 == lettersCount2:
        for letter in firstS:
            if letter in secondS:
                count1 = 0
                for symb1 in firstS:
                    if symb1 == letter:
                        count1 += 1

                count2 = 0
                for symb2 in secondS:
                    if symb2 == letter:
                        count2 += 1
                if count1 == count2:
                    return True
                else:
                    return False
            else:
                return False
    else:
        return False


twoString = input()

print(AnnoCheck(twoString))
````

Этап 1:

Этап 2:

Оформите код в виде нескольких функций:

* Зашифровывает сообщение
* Расшифровывает сообщение

````python
russian_alphabet = "а б в г д е ё ж з и й к л м н о п р с т у ф х ц ч ш щ ъ ы ь э ю я"
russian_alphabet = russian_alphabet.split()

def CezarCode():
    text = input("Введите текст, который нужно зашифровать: ")
    phase = int(input("Введите сдвиг: "))
    newText = ""
    for letter in text:
        if letter.lower() in russian_alphabet:
            letIndex = 0
            for symbol in russian_alphabet:
                if letter == symbol:
                    newLetter = russian_alphabet[(letIndex + phase) % 33]
                    newText += newLetter
                    break
                else:
                    letIndex += 1
        elif letter == ' ':
            newText += letter
    return newText

def CezarDecode():
    text = input("Введите текст, который нужно расшифровать: ")
    phase = int(input("Введите сдвиг: "))
    newText = ""
    for letter in text:
        if letter.lower() in russian_alphabet:
            letIndex = 0
            for symbol in russian_alphabet:
                if letter == symbol:
                    newLetter = russian_alphabet[letIndex - phase]
                    newText += newLetter
                    break
                else:
                    letIndex += 1
        elif letter == ' ':
            newText += letter
    return newText


print(CezarCode())
print(CezarDecode())
````
