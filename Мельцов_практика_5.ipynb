# **Задание 2**




Дан список, состоящий из целых чисел, необходимо написать функцию считающую сумму всех положительных четных чисел списка

Ввод:

```
1, 2, 3, 4, 5, 6, 7, 8, 9
```

Вывод:


```
20
```

***Запрещено:***

*   Использование готовых функций для суммирования чисел


````python
string_list = input()

def positive_sum(str_list):
  amount = 0
  num_list = str_list.split(", ")
  for number in num_list:
    if int(number) > 0 and int(number) % 2 == 0:
      amount += int(number)
  return amount

print(positive_sum(string_list))
````

# **Задание 4**




Дан словарь, необходимо написать функцию меняющую ключ и значение местами

Дано:


```
{"a": 1, "b": 2, "c": 3}
```

Вывод:

```
{1: 'a', 2: 'b', 3: 'c'}
```



````python
start_dict = {"a": 1, "b": 2, "c": 3, "d": 4}

def swap_dict(some_dict):
  new_dict = {}
  dict_keys = some_dict.keys()
  dict_values = some_dict.values()

  for key in dict_keys:
    new_dict[some_dict[key]] = key

  return new_dict



print(swap_dict(start_dict))
````

**Критерии**


*   Словарь необходимо отсортировать по убыванию количества элементов в списке.
*   Подсчет элементов должен быть реализован в отдельной функции
*   Сортировка пары `ключ:значение` должна быть реализована также в виде отдельной функции




Дано:
```
['apple','banana','orange','apple','apple','banana']
```


Вывод:
```
{'apple':3, 'banana': 2, 'orange': 1}
```

***Запрещено:***

*   Использование готовых функций для сортировки словаря
*   Использование готовых функций для подсчета элементов



````python
def sort_list(some_list):
  list_length =  0
  for words in some_list:
    list_length += 1

    sorted_list = []

  for i in range(list_length):
    biggest_word = some_list[0]
    for word in some_list:
      if len(word) > len(biggest_word):
        biggest_word = word
    sorted_list.append(biggest_word)
    some_list.remove(biggest_word)
  return sorted_list

def create_dict(some_list):
  new_dict = {}
  for element in some_list:
    new_dict[element] = len(element)
  return new_dict

input_list = input().split()
sorted_list = sort_list(input_list)
sorted_dict = create_dict(sorted_list)

print(sorted_dict)



````

Описание: Создать программу на Python для хранения и управления отзывами о предметах учебного курса. Программа должна позволять пользователям добавлять, просматривать и удалять отзывы, а также вычислять средний балл по заданному предмету.

**Функционал:**

*   Добавление отзыва и оценки:
   *   Пользователь может ввести название предмета, оценку (от 1 до 5) и текст отзыва.
   *   Отзывы должны храниться в структуре данных (например, словаре), где ключом будет название предмета, а значением - список отзывов (каждый отзыв может хранить оценку и комментарий).
*   Просмотр отзывов и оценок:
   *   Пользователь может запросить отзывы для указанного предмета.
   *   Если для указанного предмета есть отзывы, программа должна отобразить список всех отзывов и соответствующих оценок.
*   Удаление отзыва:
   *   Пользователь может удалить отзыв по индексу. Необходимо заранее уведомить пользователя о том, какие отзывы доступны для удаления.

*   Вычисление среднего балла по предмету:
   *   Пользователь может ввести название предмета, и программа должна вычислить и вывести средний балл по всем отзывам для этого предмета.
   *   Если отзывов нет, программа должна сообщить об этом.


**Критерии:**

*   Код должен быть оформлен в виде функций
*   Необходимо обрабатывать неправильный ввод пользователя
*   Должны быть комментарии к функциям
*   Присутсвует весь дополнительный функционал



**Опционально:**

Предлагаю вам добавить свои критерии оценки или вопросы, на которые должен ответить студент, чтобы оценить пару




````python

def make_a_feedback(rate_dict):
  subject_name = input("Введите название предмета: ")
  feedback_rate = int(input("Введите рейтинг отзыва: "))
  if 0 < feedback_rate <= 5:
    feedback_title = input("Введите текст отзыва: ")

    total_feedback = f"{feedback_rate}/5: {feedback_title}"

    if subject_name not in rate_dict:
      rate_dict[subject_name] = []
      rate_dict[subject_name].append(total_feedback)
    else:
      rate_dict[subject_name].append(total_feedback)
  else:
    print("Оценка должна быть от 1 до 5")

def view_feedbacks(rate_dict, subject_name = None):
  count = 1
  if subject_name is None:
    subject_name = input("Введите название предмета: ")
    if subject_name not in rate_dict:
      print("Предмет не найден")
    else:
      print(f"Все отзывы по предмету {subject_name}:")
      for rates in rate_dict[subject_name]:
        print(f"№{count}: {rates}")
        count += 1
      print("Средняя оценка: ", find_average(rate_dict, subject_name))
  else:
    print(f"Все отзывы по предмету {subject_name}:")
    for rates in rate_dict[subject_name]:
      print(f"№{count}: {rates}")
      count += 1
    print("Средняя оценка: ", find_average(rate_dict, subject_name))

def find_average(rate_dict, subject_name):
  rates_count = 0
  rates_amount = 0
  for rate in rate_dict[subject_name]:
    rates_amount += int(rate[0])
    rates_count += 1
  if rates_count != 0:
    return round((rates_amount / rates_count), 2)
  else:
    return 0

def delete_feedback(rate_dict):

  feedbacks_amount = 0
  subject_name = input("Введите название предмета: ")
  for feedbacks in rate_dict[subject_name]:
    feedbacks_amount += 1
  view_feedbacks(rate_dict, subject_name)
  rate_number = int(input("Введите номер отзыва, который нужно удалить: "))
  if 1 <= rate_number <= feedbacks_amount:
    rate_dict[subject_name].remove(rate_dict[subject_name][rate_number - 1])
    print("Отзыв успешно удалён")
  else:
    print("Отзыв с таким номером не существует")


test_dict = {}

print("Доступные команды:")
print("   Создать отзыв")
print("   Посмотреть отзывы")
print("   Удалить отзыв")
print("   Выйти из системы")

while True:
  print()
  action = input("...")
  if action == "Выйти из системы":
    break
  elif action == "Создать отзыв":
    make_a_feedback(test_dict)
  elif action == "Посмотреть отзывы":
    view_feedbacks(test_dict)
  elif action == "Удалить отзыв":
    delete_feedback(test_dict)
  else:
    print("Неизвестная команда")
````
