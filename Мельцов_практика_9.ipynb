# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def find_uniq(list_1, list_2):
  two_lists = list_1 + list_2
  uniq_elems = set(two_lists)
  return uniq_elems

a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

print(find_uniq(a, b))
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
def adopt_pet(pets_dict):
  pet_name = input('Введите имя питомца: ')
  pet_type = input('Введите вид питомца: ')
  pet_age = input('Введите возраст питомца: ')

  while True:
    if pet_name in pets_dict:
      print('У вас уже есть питомец с таким именем')
      continue
    else:
      pets_dict[pet_name] = {'type': pet_type, 'age': pet_age, 'hunger': 100, 'happines': 100, 'need_sleep': 0}
      break

def pet_action(pets_dict):
  while True:
    pet_name = input('Введите имя питомца: ')
    if pet_name not in pets_dict:
      print('У вас нет питомца с таким именем')
    else:
      print('Доступные команды: ')
      print('1: Поиграть')
      print('2: Покормить')
      print('3: Положить спать')
      print('4: Выйти')

      action = input()
      if action == '1':
        print(f'Вы поиграли с питомцем: {pet_name}')
        pets_dict[pet_name]['happines'] += 20
        if pets_dict[pet_name]['happines'] > 100:
          print(f'{pet_name} вне себя от счастья!')
        else:
          print('Уровень счастья:', pets_dict[pet_name]['happines'])

      elif action == '2':
        print(f'Вы покормили питомца: {pet_name}')
        pets_dict[pet_name]['hunger'] += 20
        if pets_dict[pet_name]['hunger'] > 100:
          print(f'{pet_name} объелся!')
        else:
          print('Уровень сытости:', pets_dict[pet_name]['hunger'])

      elif action == '3':
        print(f'Вы уложили спать питомца: {pet_name}')
        pets_dict[pet_name]['need_sleep'] -= 5
        if pets_dict[pet_name]['need_sleep'] < 0:
          print(f'{pet_name} выспался!')
        else:
          print('Желание спать:', pets_dict[pet_name]['need_sleep'])

      elif action == '4':
        break

      else:
        print('Такой команды нет')



def watch_info(pets_dict):
    pet_name = input('Введите имя питомца: ')
    if pet_name not in pets_dict:
      print('У вас нет питомца с таким именем')
      return '404'

    print(f'Имя: {pet_name}')
    print('Возраст:', pets_dict[pet_name]['age'])
    print('Вид:', pets_dict[pet_name]['type'])
    print('Сытость:', pets_dict[pet_name]['hunger'])
    print('Счастье:', pets_dict[pet_name]['happines'])
    print('Потребность в сне:', pets_dict[pet_name]['need_sleep'])



pets = {}
print('Доступные команды: ')
print('1: Приютить питомца')
print('2: Информация о питомце')
print('3: Взаимодействие с питомцем')
print('4: Выход')

while True:
  action = input('Введите команду: ')
  if action == '1':
    adopt_pet(pets)
  elif action == '2':
    watch_info(pets)
  elif action == '3':
    pet_action(pets)
  elif action == '4':
    break
  else:
    print('Неизвестная команда')


  for pet in pets:
    pets[pet]['hunger'] -= 5
    pets[pet]['happines'] -= 5
    pets[pet]['need_sleep'] += 1



````
